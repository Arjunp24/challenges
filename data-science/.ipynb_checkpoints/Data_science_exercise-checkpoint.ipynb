# Library Imports  

All necesary libraries used for this challenge are imported here.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
%matplotlib inline
```

# UNDERSTANDING THE DATASET  
<br><br>
After reading the csv, we understand that there are four columns, each column representing the four kinds of time series data and the dataset is indexed with its corresponding time. The time also seems sorted in order.  
There are no missing values or nan present in the dataset so, there isn't any need to perform imputation.
<br>

```python
data = pd.read_csv('./machine_0.csv',index_col=0)
print(data.head(),'\n')
print(data.info(),'\n')
print(data.describe(),'\n')
print(data.isnull().sum())
```

# DATA VISUALISATION  
<br><br>
We are given 3 modes: normal, faulty and failed. As the dataset does not explicitly tell us anything about these modes, we will have to visualise it.  
We know that the normal mode behaves predictably, faulty mode is a little bit off and failed is around 0.  
<br>
From the useful visualization provided, we can immediately spot a line along the x-axis. This behaviour is more likely to represent the failed mode.  
Other than that there are too many lines and some lines get overlapped by others. So, make it look much clearer, I'm plotting a scatterplot against each kind of time series data.  
From these scatterplots we observe a pattern going along the x-axis and two more patterns appearing between 200 < y < 300 and -200 < y < -300.  
<br>
At the beginning of the x-axis, we see the data points behave in pretty stable fashion and then the behavior visibly changes before it collapses into a straigth line along the x-axis.  
From this behaviour, we can infer that the stable behavior observed represents the normal mode and the slightly different behaviour is the faulty mode.  
<br>
This inference also makes sense as when a machine is brand new, it functions normally and then after a point in time, it becomes faulty and finally after more wear and tear, it starts to fail.  
This also means that the other two patterns we observed tell us that they are noise as they really do not contribute in describing the behavior of the machine.

```python
plt.figure(figsize=(10,10))

plt.plot(range(len(data)), data)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,20))
sns.scatterplot(x=range(len(data)),y = data['0'], ax=ax[0][0])
sns.scatterplot(x=range(len(data)),y = data['1'], ax=ax[0][1])
sns.scatterplot(x=range(len(data)),y = data['2'], ax=ax[1][0])
sns.scatterplot(x=range(len(data)),y = data['3'], ax=ax[1][1])
```

# REMOVING OUTLIERS  
<br><br>
The outliers detected need to be removed. We can see that they appear approximately when 200 < y < 300 and -200 < y < -300. After removing the outliers, we find that there are around 2450 records out of the original 3000 records. This means we have not lost a huge chunk of our data and we can proceed further.  
<br>
Plotting the dataset now after removing the outliers makes the visualization much easier. The initial line plot which was provided now looks clearer and we can easily identify the 3 modes from plot.

```python
new_data = data[(data['0'] < 200) & (data['0'] > -200) & 
            (data['1'] < 200) & (data['1'] > -200) &
            (data['2'] < 200) & (data['2'] > -200) &
            (data['3'] < 200) & (data['3'] > -200)]
print(new_data)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,20))
plt.figure()
sns.scatterplot(x=range(len(new_data)),y = new_data['0'], ax=ax[0][0])
sns.scatterplot(x=range(len(new_data)),y = new_data['1'], ax=ax[0][1])
sns.scatterplot(x=range(len(new_data)),y = new_data['2'], ax=ax[1][0])
sns.scatterplot(x=range(len(new_data)),y = new_data['3'], ax=ax[1][1])
plt.show()

plt.plot(range(len(new_data)), new_data)
plt.show()
```

# NEW FEATURE: AGE  
<br><br>
The objective of the challenge is to identify the times a machine will enter faulty or failed mode. The time is present as the index of the dataset. This does not let us use this data in anyway. The visualization we have arrived at also does not specify the time as we plotted the data points against the length of the dataset.  
<br>
This calls for a new feature which involves the time. But, the time present in the index is specific to that machine, say 2019-01-01 00:00:00.000000000, it only represents one time machine_0 worked at. This time may or may not be a time another machine would work at. Hence, we need to represent this time in a more generic format so all machines can be compared using this format. As a result, I have calculated the age of the machine to solve this problem.  
<br>
To do this, we calculate the timestamp for every time. We assume the timestamp present in the first row is when the machine was first started to being in use, which means its age is zero. Similarly, every other row gets its age calculated by subtracting the timestamp of the first row from the timestamp of its corresponding row. By doing this, we can now generalize the times for a machine to identify which mode it is in.  
<br>
NOTE: I had also scaled, normalized and performed feature extraction using PCA but, the data points but upon plotting it did not help solve the problem in hand. Therefore, I went against it and stuck to the given dataset.

```python
new_data['time'] = pd.to_datetime(new_data.index, unit='ns')
index = new_data.index
new_data['time_stamp'] =  new_data.time.values.astype(np.int64) // 10 ** 9
new_data = new_data.drop('time', axis= 1)
print(new_data)

new_data['age_in_ms'] = new_data['time_stamp'] - new_data.loc[new_data.index[0],'time_stamp']
new_data = new_data.drop('time_stamp', axis= 1)
print(new_data)
```

# MACHINE LEARNING  
<br><br>
Our dataset now has the four kinds of time series data and the age. We are already aware that this dataset can be categorised into 3 categories: normal, faulty, failed. But the dataset is not labelled which kind of makes this problem a mix of unsupervised and supervised learning. Therefore, we will first try clustering the dataset such that the clusters cover the three modes appropriately.

```python
kmeans = KMeans(n_clusters=13, max_iter=100, random_state=0).fit_predict(new_data)
new_data['cluster'] = kmeans

plt.figure(figsize=(15,15))
sns.scatterplot(x = new_data['age_in_ms'], y = new_data['1'], hue = new_data['cluster'], 
                palette=['black', 'red', 'blue', 'orange', 'green', 'brown', 'yellow', 'gray'
                         ,'violet', 'pink', 'purple', 'magenta', 'maroon'])
plt.show()
```

# RESULTS  
<br><br>
After performing clustering analysis with KMeans, Spectral, Agglomerative, Hierarchical and DBSCAN clustering, I got the right clustering results with KMeans, having 13 number of clusters. 13 doesn't seem right as we have only 3 categories but the way the 13 clusters are made can be visualised as 3 clusters that cover our 3 categories. From the clustering plot, we can see the cluster 6 contains the data under normal mode. Cluster 10 contains the data under faulty mode and the rest of the clusters contain the data under failed mode.  
<br>
Therefore, we can represent the data points from these clusters which will pinpoint the times in each category.  
<br>
NOTE: Even though I have set a random_state for the model, the nature of KMeans algorithm is not deterministic in nature. As a result, the clusters 6 and 10 that were chosen may vary when you run it. But what's stable is the fact this model covers normal mode with one cluster, faulty mode with cluster and failed mode with the remaining clusters.

```python
normal_list = []
faulty_list = []
failed_list = []
for i in range(len(new_data)):
    if new_data.iloc[i, 5] == 6:
        normal_list.append(new_data.index[i])
for i in range(len(new_data)):
    if new_data.iloc[i, 5] == 10:
        faulty_list.append(new_data.index[i])
for i in range(len(new_data)):
    if new_data.iloc[i, 5] in [0,1,2,3,4,5,7,8,9,11,12]:
        failed_list.append(new_data.index[i])

df = pd.DataFrame({'Normal': pd.Series(normal_list), 'Faulty': pd.Series(faulty_list), 'Failed': pd.Series(failed_list)})
print(df)
```
